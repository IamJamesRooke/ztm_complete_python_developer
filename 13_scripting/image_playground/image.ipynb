# Image Playground

Imports

```python
from PIL import Image
import os
```

Opening Image

```python
image_path = './Pokedex/pikachu.jpg'

try:
    img = Image.open(image_path)
    print(f'Successfully opened image: {img.size}')
except FileNotFoundError:
    print(f'File not found - please check that the file exists at {image_path}')
except Exception as e:
    print(f'Error opening image: {e}')
```

Showing Image

```python
display(img)
```

Filters

```python
from PIL import ImageFilter
filtered_img = img.filter(ImageFilter.BLUR)
display(filtered_img)
filtered_img.save('blur.png')
```

Smooth

```python
filtered_img2 = img.filter(ImageFilter.SMOOTH)
display(filtered_img2)
```

Sharpen

```python
filtered_img3 = img.filter(ImageFilter.SHARPEN)
display(filtered_img3)
```

Greyscale

```python
filtered_img = img.convert('L')
display(filtered_img)
```
